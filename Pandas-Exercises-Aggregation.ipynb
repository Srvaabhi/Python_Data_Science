### Step 4. Discover what is the mean age per occupation

```python
Mean_Age = users.groupby(by="occupation")[["age"]].mean().reset_index()
Mean_Age
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users = users.assign(Age_encoding = pd.Series(np.where(users.gender == "M",1,0)))
Male_data = users.loc[users.gender=="M"]
Male_data.head()
```

```python
Male_Group  = Male_data.groupby(by="occupation")[["Age_encoding"]].count().add_prefix("Male").reset_index()
Male_Group.head()
```

```python
Total_Workers_Count = users.groupby(by=["occupation"])[["Age_encoding"]].count()
Total_Workers_Count.head()
```

```python
L = pd.merge(left = Male_Group,right=Total_Workers_Count,on="occupation",how="inner")
L.head()
```

```python
L = L.assign(Male_Ratio = L.MaleAge_encoding/L.Age_encoding)
L
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
resultant
```

```python
Minimum_age = users.groupby(by = "occupation")[["age"]].min().add_prefix("Minimum_").reset_index()
Minimum_age.head()
```

```python
Maximum_age = users.groupby(by = "occupation")[["age"]].max().add_prefix("Maximum_").reset_index(drop = True)
Maximum_age.head()
```

```python
resultant = pd.concat([Minimum_age,Maximum_age],axis =1)
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.head()
```

```python
GenderWise_Count = users.groupby(by=["occupation","gender"])[["gender"]].count().add_prefix("CountOf_").reset_index()
```

```python
GenderWise_Count.head()
```

```python
gg = GenderWise_Count.groupby(by = ["occupation"])[["CountOf_gender"]].sum().add_prefix("Total").reset_index()
```

```python
gg.head()
```

```python
v = pd.merge(left =GenderWise_Count,right=gg,on="occupation",how="inner")
```
