### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
#t = online_rt.Country[-online_rt.Country.duplicated()]
```

```python
online_rt1 = online_rt.loc[-(online_rt.Country == "United Kingdom")]
online_rt1
```

```python
hist = online_rt1.groupby(by = ["Country"])[["Quantity"]].count().add_prefix("CountOf_").reset_index()
hist = hist.sort_values(by = "CountOf_Quantity",ascending = False)
q = hist.head(10)
q 
```

```python
q.CountOf_Quantity.plot(kind = "hist", bins = 20, color = "indianred", edgecolor = "black")
plt.xlabel("Quantity Ordered by Top 10 Country")
plt.title("Histogram of Quantity from Online_rt1")
plt.show()
```

```python
w = plt.hist(q.CountOf_Quantity,bins = 20,color = "skyblue", edgecolor = "black")
plt.xlabel("Quantity Ordered by Top 10 Country")
plt.title("Histogram of Quantity from Online_rt1 ")
plt.show()
```
