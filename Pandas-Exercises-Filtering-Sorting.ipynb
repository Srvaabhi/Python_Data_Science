### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
chipo.item_price = chipo.item_price.str.replace("$","")
chipo.item_price = pd.to_numeric(chipo.item_price,errors = "coerce")
```

```python
sub1 = chipo.loc[chipo.item_price>10].reset_index(drop = True)
sub1
```

### Step 5. What is the price of each item? 

print a data frame with only two columns `item_name` and `item_price`

```python
sub2 = chipo.loc[:,["item_name","item_price"]].reset_index(drop=True)
sub2 = sub2.loc[-sub2.duplicated(),]
sub2
```

### Step 6. Sort by the name of the item

```python
sort2 = chipo.sort_values(by = "item_name")
sort2
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
sort1 = chipo.sort_values(by = "item_price",ascending = False)
sort1
#15 is the quantity of the most expensive item ordered
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
filter1 = chipo.loc[chipo.item_name == "Veggie Salad Bowl"].reset_index(drop = True)
filter1
#Veggie Salad Bowl is ordered 18 times
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
filter2 = chipo.loc[(chipo.quantity > 1) & (chipo.item_name == "Canned Soda")].reset_index(drop = True)
filter2
# 20 times People have ordered more than one canned soda
```
